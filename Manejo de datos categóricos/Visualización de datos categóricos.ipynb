## 3. Visualización de datos categóricos

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("datos\lasvegas_tripadvisor.csv") # Importar data set
```

### Diagrama de cajas

```python
# Configurar el tamaño de fuente
sns.set(font_scale=1.25)

# Configurar el color de fondo de los gráficos
sns.set_style("whitegrid")

# Crear un boxplot
sns.catplot(x="Traveler type", y="Helpful votes", data=df, kind="box")

plt.show()
```

## Diagrama de barras

```python
# Order las categorias y gráficar

df["Traveler type"] = df["Traveler type"].astype("category")
df["Traveler type"] = df["Traveler type"].cat.set_categories(['Business','Solo','Friends',
                                                         'Families', 'Couples'], ordered=True)
```

```python
sns.catplot(x="Traveler type", y="Score", data=df, kind="bar")
plt.xticks(rotation=40) # Cambiar la orientación de los ejes
plt.show()
```

```python
# Gráfico categórico que presenta el conteo de puntaje a travéz del tipo de viajero
# Los argumentos heigh

sns.catplot(x = "Score", data=df, kind="count", hue="Traveler type", height=5, col_, aspect=2)
plt.show()
```

```python
# Crear un gráfico diferenciado por columnas y colores. Wrap indica cada cuantas columnas se da un salto

ax = sns.catplot( x="Review weekday", col="Period of stay", kind="count", col_wrap=2, data=df)
ax.set_xticklabels(rotation=30) # Rotar las etiquetas de los ejes
plt.subplots_adjust(top=0.9)
plt.show()
```

### Diagrama de intervalos (point plots)


```python
# Join elimina la línea entre intervalos

sns.catplot(x="Score", y="Review weekday", data=df, kind="point", capsize=0.1, join=False) 
plt.title("Intervalos para subgrupos")
plt.show()
```
